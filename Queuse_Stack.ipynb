# Queue
Like a stack, the queue is a linear data structure that stores items in a First In First Out (FIFO) manner. <br> With a queue, the least recently added item is removed first. <br> A good example of a queue is any queue of consumers for a resource where the consumer that came first is served first.



adding direction <<<   last element..... third element, second element, first element added

Operations associated with queue are: 

* **Enqueue**: <span style="color:green; font-weight:bold;">Adds</span> an item to the queue. If the queue is full, then it is said to be an Overflow condition – Time Complexity : O(1)
* **Dequeue**: <span style="color:red; font-weight:bold;">Removes</span> an item from the queue. The items are popped in the same order in which they are pushed. <br >If the queue is empty, then it is said to be an Underflowcondition – Time Complexity : O(1)
* **Front**: <span style="color:green; font-weight:bold;">Get (read)</span> the front item from queue – Time Complexity : O(1)
* **Rear**:   <span style="color:green; font-weight:bold;">  Get (read)</span> the last item from queue – Time Complexity : O(1)
